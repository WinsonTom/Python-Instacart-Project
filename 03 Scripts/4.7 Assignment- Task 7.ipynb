```python
# Import libraries

import pandas as pd
import numpy as np
import os
```

```python
#Import Orders products merged
ords_prods_merge = pd.read_pickle(r'/Users/winsontom/Desktop/10-2023 Instacart Basket Analysis/02 Data /Prepared Data/orders_products_merged.pkl')
```

```python
#Only work with 1 million rows per task

df = ords_prods_merge[:1000000]
```

```python
df
```

```python
#Created function to create price label

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Creating a price label column; data put into 1 which is rows. 

df['price_range'] = df.apply(price_label, axis=1)
```

```python
#Value count of the  different ranges, drop it if NA
#Mid Range products are most common

df['price_range'].value_counts(dropna = False)
```

```python
#Maximum price

df['prices'].max()
```

```python
#Using Loc to define the high range products. Locate - prices column - value more than 15, then in price range loc put it in as high range product
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
#Defining mid range product values for DF
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
##Defining low range product values for DF
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#Apply the LOC method to full data not just 1 million rows

ords_prods_merge.loc[ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
#Apply the LOC method to full data - Mid range
ords_prods_merge.loc[(ords_prods_merge['prices'] <= 15) & (ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
#Apply the LOC method to full data - Low range

ords_prods_merge.loc[ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#Tracking which day of the week has the most orders. 0 = Saturday. 

ords_prods_merge['order_dow'].value_counts(dropna = False)
```

```python
#New column that identifies if it is busiest, least busy, regularly busy

result = []

for value in ords_prods_merge["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
ords_prods_merge['busiest_day'] = result
```

```python
#Counting what occurs the most, 
ords_prods_merge['busiest_day'].value_counts(dropna = False)
```
