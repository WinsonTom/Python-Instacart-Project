```python
#Step 4 - Creating a loyalty flag for existing customers using the transform() and loc() functions. Full DF
#First step is creating this "max order" column to identify what level of loyalty are they
#Transform function

DF['max_order'] = DF.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#Step 4 - Creating a loyalty flag for existing customers using the transform() and loc() functions. Full DF

#Create loyalty_flag column
#Loc create inputs new column
#Conditions for new column to define a loyal customer, regular customer, and new customer. 
#Locate max orders in DF. If its more than 40, fill in loyalty_flag column with "Loyal Customer"

DF.loc[DF['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
#Locate max orders in DF. If its more than 10 and less than or equal to 40
#Fill in loyalty_flag column with "Regular Customer"
#Note: Bracket column setup is very important

DF.loc[(DF['max_order'] <= 40) & (DF['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
#Locate max orders in DF. If its less than or equal to 10 , then fill in the Loyalty_Flag Column with "New Customer"

DF.loc[DF['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Step 5: Identify differences in spending habits. 
#Using the loyalty flag you created compare basic statistics of the product prices for each category 
#How do prices of products differ between loyal vs regular/new customer.
```

```python
#AGG function to look at loyalty levels
#Mean - average prices for loyal, new, regular levels
#Table shows differences in spending by customer loyalty. 
#As the customers shop more often, the mean / average prices decrease. 
#New customers spend the most. Regular customers spend less. Loyal Customers spend the least.

DF.groupby('loyalty_flag').agg({'prices': ['mean','min','max']})
```

```python
#Step 6 - Target different types of spenders
#Spending flag for each user based on the average price across all their orders
#Mean<10 = Low Spender
#Mean>10 = High Spender

DF['Average_Price_Of_Orders'] = 'Values'
DF[['user_id', 'prices', 'Average_Price_Of_Orders']].head(150)
```

```python
#Seperated by the user ids , find the average of these prices
#AGG function
#This is a table of the mean prices, however we want the information to actually fill out the column Average_Price_of_Orders


DF.groupby('user_id').agg({'prices': ['mean']})
```

```python
#Filled out the columns with Transform(np.mean) function. Not np.max
#Result is Average_Price_of_Orders filled out. It contains the Avg price for each user ID.
#Formula means "Dataframe column of Average Price of Orders is Dataframe grouped by the user ID, fill the value with prices-average"

DF['Average_Price_Of_Orders'] = DF.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Check the DF to see if the column is filled out properly
DF.head(15)
```

```python
DF.loc[DF['Average_Price_Of_Orders'] < 10 , 'Spender_Type'] = 'Low Spender'
```

```python
DF.loc[DF['Average_Price_Of_Orders'] >= 10 , 'Spender_Type'] = 'High Spender'
```
